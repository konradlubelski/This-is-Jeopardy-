```python
#Load data

data = pd.read_csv('jeopardy.csv')
```

```python
#Column names

column_name_list = list(data.columns)
column_name_list
```

```python
#Rename columns

data.rename(columns={' Air Date':'Air Date', ' Round': 'Round', ' Category': 'Category'
                      , ' Value': 'Value', ' Question': 'Question', ' Answer': 'Answer'}, inplace = True)


list(data.columns)
```

```python
#Function that finds questions containing all given words from the list

def find_question_with_words_in_list(lst, df):
  len_lst = len(lst)
  result_df = pd.DataFrame(columns = data.columns)
  for ind in df.index:
    sum_words_in_question = 0
    for word in lst:
      if word in df.Question[ind]:
        sum_words_in_question += 1
    if sum_words_in_question == len_lst:
      result_df = result_df.append(df.iloc[ind])
  return result_df


a = find_question_with_words_in_list(['King', 'England'], data)

a
      
```

```python
#Function that finds questions containing all given words from the list - 2nd version

def find_question_with_words_in_list_2nd(lst, df):
  len_lst = len(lst)
  result_df = pd.DataFrame(columns = data.columns)
  for ind in df.index:
    sum_words_in_question = 0
    for word in lst:
      if ' ' + word + ' ' in df.Question[ind] or ' ' + word.lower() + ' ' in df.Question[ind]:
        sum_words_in_question += 1
    if sum_words_in_question == len_lst:
      result_df = result_df.append(df.iloc[ind])
  return result_df


find_question_with_words_in_list_2nd(['King', 'England'], data)
```

```python
import datetime

#For delete sign $ from first place of string Value column, only 1-time use 
for ind in data.Value.index:
  data.Value[ind] = data.Value[ind][1:]

#Change 'one' word to string '1' for further changing in float type
data.Value.replace('one', '1', inplace = True)

#Change from original , grupp point to empty space
data.Value = data.Value.str.replace(',','')

#Converting column Value to float format
convert_dict = {'Value': float}

data = data.astype(convert_dict)

#Converting column Air Date to datetime format
data['Air Date'] = pd.to_datetime(data['Air Date'])

#Mean for subset of data
subset = find_question_with_words_in_list(['King', 'England'], data)

subset.Value.mean()
```
